```python
import numpy as np
# Quadratic training set + noise
np.random.seed(42)
m = 200
X = np.random.rand(m, 1)
y = 4 * (X - 0.5) ** 2
Y = y + np.random.randn(m, 1) / 10
```

1. Try to plot the datase

```python
import matplotlib. pylab as plt

plt.plot(X, Y, "bo")
plt.xlabel("$x_{1}$", fontsize=18)
plt.ylabel("$y$", fontsize=18, rotation=0)
plt.show()
```

2 Construct a decision tree model with max_depth =2

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor(max_depth=2)
tree_reg.fit(X,Y)
```

 3 Visualize the regression tree¶

```python
from sklearn.tree import export_graphviz
export_graphviz(tree_reg,
                out_file="tree_reg.dot",
                feature_names=["X"],
                class_names=["Y"],
                rounded=True,
                filled=True,
               )
# Convert dot file to png file.
!dot -Tpng tree_reg.dot -o tree_reg.png
from IPython.display import Image
Image(filename='tree_reg.png') 
```

4 Plot this regression tree (Tip: try many values for x (e.g. np.linspace(min, max, noOfPoints))



```python
Xs = np.linspace(0, 1, 200).reshape(-1, 1)
                                                        
Ys = tree_reg.predict(Xs)
plt.plot(Xs, Ys, "ro")
plt.plot(X, Y, "bo")
plt.xlabel("X", fontsize=18)
plt.ylabel("Y", fontsize=18, rotation=0)
plt.show()
```

 5 Plot the decision boundaries of max_depth=3 and max_depth=4 regression trees (also try min_samples_leaf=10)

```python
tree_reg = DecisionTreeRegressor(max_depth=3)
tree_reg.fit(X,Y)
Xs = np.linspace(0, 1, 100).reshape(-1, 1)
Ys = tree_reg.predict(Xs)
plt.plot(Xs, Ys, "ro")
plt.plot(X, Y, "bo")
plt.xlabel("X$", fontsize=18)
plt.ylabel("$Y$", fontsize=18, rotation=0)
plt.show()
```

```python
tree_reg = DecisionTreeRegressor(max_depth=4)
tree_reg.fit(X,Y)
Xs = np.linspace(0, 1, 100).reshape(-1, 1)
Ys = tree_reg.predict(Xs)
plt.plot(Xs, Ys, "ro")
plt.plot(X, Y, "bo")
plt.xlabel("$X$", fontsize=18)
plt.ylabel("$Y$", fontsize=18, rotation=0)
plt.show()
```

```python
tree_reg = DecisionTreeRegressor(max_depth=4,min_samples_leaf=10)
tree_reg.fit(X,Y)
Xs = np.linspace(0, 1, 100).reshape(-1, 1)
Ys = tree_reg.predict(Xs)
plt.plot(Xs, Ys, "ro")
plt.plot(X, Y, "bo")
plt.xlabel("$X$", fontsize=18)
plt.ylabel("$Y$", fontsize=18, rotation=0)
plt.show()
```
